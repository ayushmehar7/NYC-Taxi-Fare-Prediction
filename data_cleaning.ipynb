```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
taxi = pd.read_csv("NYCTaxiFares.csv")
```

```python
taxi.head()
```

```python
taxi['pickup_datetime'] = pd.to_datetime(taxi['pickup_datetime'])
```

```python
taxi['pickup_datetime'] = pd.to_datetime(taxi['pickup_datetime'] - pd.Timedelta(hours = 4))
```

```python
taxi.head()
```

```python
taxi['pickup_hour'] = taxi['pickup_datetime'].apply(lambda x : x.hour)
```

```python
taxi = taxi.drop('pickip_hour',axis = 1)
taxi.head()
```

```python
day_dict = {0:'Mon',1:'Tue',2:'Wed',3:'Thurs',4:'Fri',5:'Sat',6:'Sun'}
taxi['day_of_week'] = taxi['pickup_datetime'].apply(lambda x : day_dict[x.dayofweek])
```

```python
taxi.head()
```

```python
taxi = taxi.drop('fare_class',axis = 1)
```

```python
def amOrpm(x):
    if x.hour < 12:
        return 'AM'
    else:
        return 'PM'
```

```python
taxi['am_or_pm'] = taxi['pickup_datetime'].apply(amOrpm)
```

```python
taxi.head()
```

```python
from math import radians, cos, sin, asin, sqrt 
def distance(lat1, lat2, lon1, lon2):  
    lon1 = radians(lon1) 
    lon2 = radians(lon2) 
    lat1 = radians(lat1) 
    lat2 = radians(lat2) 
    dlon = lon2 - lon1  
    dlat = lat2 - lat1 
    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * asin(sqrt(a))  
    r = 6371
    return(c * r) 
```

```python
dis = []
dat = taxi['pickup_longitude pickup_latitude dropoff_longitude dropoff_latitude'.split()].values
for x in dat:
    lat1 = float(x[1])
    lat2 = float(x[3])
    log1 = float(x[0])
    log2 = float(x[2])
    dis.append(distance(lat1,lat2,log1,log2))
taxi['journey_distance'] = dis
```

```python
taxi.head()
```

```python
taxi = taxi['journey_distance am_or_pm day_of_week pickup_hour passenger_count fare_amount'.split()]
```
